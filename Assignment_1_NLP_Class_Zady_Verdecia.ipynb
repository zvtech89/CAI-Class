```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print(f"{'Token':<15} | {'Stemmed':<15} | {'Lemmatized':<15}")
print("-" * 45)

for token in cleaned_tokens:
    stem = stemmer.stem(token)
    lemma = lemmatizer.lemmatize(token)
    print(f"{token:<15} | {stem:<15} | {lemma:<15}")
```

```python
# POS Tagging
pos_tags = nltk.pos_tag(cleaned_tokens)
print("\nPOS Tags:", pos_tags)

# Named Entity Recognition (NER)
# We use the original tokens for NER because capitalization matters for identifying names!
ner_tags = nltk.pos_tag(word_tokenize(my_text))
chunks = nltk.ne_chunk(ner_tags)

print("\nIdentified Entities:")
for chunk in chunks:
    if hasattr(chunk, 'label'):
        print(f"{' '.join(c[0] for c in chunk):<20} | Type: {chunk.label()}")
```
